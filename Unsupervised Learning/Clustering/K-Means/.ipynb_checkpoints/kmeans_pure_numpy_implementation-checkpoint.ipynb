```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
points = np.vstack((
    np.random.normal([2,2], 0.5, (50,2)),
    np.random.normal([6,6], 0.5, (50,2)),
    np.random.normal([10,2], 0.5, (50,2))
))

K = 3
max_iters = 150

indices = np.random.choice(points.shape[0], K, replace=False)
centroids = points[indices]

for i in range(max_iters):
    distances = np.linalg.norm(points[:, np.newaxis] - centroids, axis=2)
    labels = np.argmin(distances, axis=1)

    new_centroids = np.array([points[labels == k].mean(axis=0) for k in range(K)])

    if np.allclose(centroids, new_centroids):
        break
    centroids = new_centroids

for k in range(K):
    cluster_points = points[labels == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {k}')

plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, marker='X', label='Centroids')
plt.title('"K-Means from Scratch"')
plt.legend()
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set seed for reproducibility
np.random.seed(42)

# Generate the sample data - 3 distinct clusters
points = np.vstack((
    np.random.normal([2, 2], 0.5, (50, 2)),  # Cluster 1
    np.random.normal([6, 6], 0.5, (50, 2)),  # Cluster 2
    np.random.normal([10, 2], 0.5, (50, 2))  # Cluster 3
))

# Plot the original data to verify it looks correct
plt.figure(figsize=(10, 6))
plt.scatter(points[:50, 0], points[:50, 1], label='True Cluster 1')
plt.scatter(points[50:100, 0], points[50:100, 1], label='True Cluster 2')
plt.scatter(points[100:, 0], points[100:, 1], label='True Cluster 3')
plt.title("Original Data Distribution")
plt.legend()
plt.grid(True)
plt.savefig('original_data.png')
plt.close()

# K-means parameters
K = 3
max_iters = 150

# Choose initial centroids
np.random.seed(42)  # Reset seed for consistent centroid selection
indices = np.random.choice(points.shape[0], K, replace=False)
centroids = points[indices].copy()
print("Initial centroids:")
print(centroids)

# Store history for visualization
centroid_history = [centroids.copy()]
label_history = []

# Run K-means algorithm
for i in range(max_iters):
    # Calculate distances between each point and each centroid
    distances = np.linalg.norm(points[:, np.newaxis] - centroids, axis=2)
    
    # Assign each point to the closest centroid
    labels = np.argmin(distances, axis=1)
    label_history.append(labels.copy())
    
    # Recalculate centroids
    new_centroids = np.array([points[labels == k].mean(axis=0) if np.sum(labels == k) > 0 
                               else centroids[k] for k in range(K)])
    
    # Check for convergence
    if np.allclose(centroids, new_centroids, rtol=1e-5, atol=1e-5):
        print(f"Converged after {i+1} iterations")
        break
        
    centroids = new_centroids.copy()
    centroid_history.append(centroids.copy())

# Final result
print("Final centroids:")
print(centroids)
print("Number of points in each cluster:", [np.sum(labels == k) for k in range(K)])

# Plot the final clustering result
plt.figure(figsize=(10, 6))
for k in range(K):
    cluster_points = points[labels == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {k}')

plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering Result")
plt.legend()
plt.grid(True)
plt.savefig('final_clustering.png')
plt.show()

# Create animation of the clustering process
fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()
    
    # If we have labels for this frame
    if frame < len(label_history):
        labels = label_history[frame]
        for k in range(K):
            cluster_points = points[labels == k]
            ax.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {k}')
    else:
        # Just plot the points for the initial frame
        ax.scatter(points[:, 0], points[:, 1], c='blue', alpha=0.5)
    
    # Plot the centroids for this frame
    if frame < len(centroid_history):
        frame_centroids = centroid_history[frame]
        ax.scatter(frame_centroids[:, 0], frame_centroids[:, 1], 
                   c='black', s=200, marker='X', label='Centroids')
    
    ax.set_title(f"K-Means Iteration {frame}")
    ax.legend()
    ax.grid(True)
    return ax,

# Create animation - reduce frames if there are too many
frames = min(30, len(centroid_history))
selected_frames = np.linspace(0, len(centroid_history)-1, frames, dtype=int)
ani = FuncAnimation(fig, update, frames=selected_frames, interval=300, blit=False)

# Save the animation
try:
    ani.save('kmeans_animation.gif', writer='pillow', fps=2)
    plt.close()
    print("Animation saved successfully")
except Exception as e:
    print(f"Could not save animation: {e}")
    
# Plot initial vs final state
plt.figure(figsize=(12, 5))

# Initial state
plt.subplot(1, 2, 1)
plt.scatter(points[:, 0], points[:, 1], c='blue', alpha=0.5)
plt.scatter(centroid_history[0][:, 0], centroid_history[0][:, 1], 
           c='black', s=200, marker='X')
plt.title("Initial State")
plt.grid(True)

# Final state
plt.subplot(1, 2, 2)
for k in range(K):
    cluster_points = points[labels == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {k}')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, marker='X')
plt.title("Final State")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('initial_vs_final.png')
plt.show()
```
