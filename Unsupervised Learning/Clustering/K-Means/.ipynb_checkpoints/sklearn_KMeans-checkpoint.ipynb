```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with 3 clusters
points = np.vstack((
    np.random.normal([2, 2], 0.5, (50, 2)),  # Cluster 1
    np.random.normal([6, 6], 0.5, (50, 2)),  # Cluster 2
    np.random.normal([10, 2], 0.5, (50, 2))  # Cluster 3
))

# Number of clusters
K = 3
# Maximum iterations for convergence
max_iters = 150

# Apply K-means clustering using sklearn
kmeans = KMeans(
    n_clusters=K,          # Number of clusters to form
    max_iter=max_iters,    # Maximum number of iterations
    n_init=10,             # Number of times to run with different centroid seeds
    random_state=42        # Random seed (for reproducibility)
)

# Fit the model to the data
kmeans.fit(points)

# Get the cluster assignments for each point
labels = kmeans.labels_

# Get the final centroid positions
centroids = kmeans.cluster_centers_

# Plot the results
plt.figure(figsize=(10, 6))

# Plot all points, colored by cluster assignment
colors = ['#4EACC5', '#FF9C34', '#4E9A06']
for i in range(K):
    cluster_points = points[labels == i]
    plt.scatter(
        cluster_points[:, 0], 
        cluster_points[:, 1], 
        c=colors[i], 
        label=f'Cluster {i+1}',
        alpha=0.7
    )

# Plot the centroids
plt.scatter(
    centroids[:, 0], 
    centroids[:, 1], 
    c='red', 
    marker='X', 
    s=200, 
    label='Centroids'
)

plt.title('K-means Clustering Results')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Print the inertia (sum of squared distances of samples to their closest centroid)
print(f"Inertia: {kmeans.inertia_:.2f}")
```
